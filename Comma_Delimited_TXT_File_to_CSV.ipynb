```python
txt_name = input("Name of your TXT File (don't include extention): ") + ".txt"
data_sep = input("How is your data seperated( ' , ' or ' ; ' or ' - '): ")

# opening the file in read mode
my_file = open(txt_name, "r")

# reading the file
data = my_file.read()

# replacing end splitting the text
# when newline ('\n') is seen.
print("Data Before: ", data)
data_into_list = data.split(data_sep)
print("Data After: ", data_into_list)
my_file.close()
```

**Converting every String Data into int (Optional)**

```python
#only works if there is no error in the dataset
#example: if the dataset ends with incomplete data like- 456,678,...,123,234, <- INCOMPLETE
#then it shows error

lst = []
for i in data_into_list:
  lst.append(int(i))
print(lst)
print(len(lst))
```

**Enter DataFrame Details according to your Usecase**

```python
#Splitting the data
c_sub = []
c_main = []

no_of_col = int(input("Enter Number of Columns:"))

for i in range(0,len(data_into_list)):
  if i % no_of_col == 0:
    c_main.append(c_sub)
    c_sub = []
    c_sub.append(data_into_list[i])
  else:
    c_sub.append(data_into_list[i])

c_main.pop(0)

# import pandas as pd
import pandas as pd

c = []
for p in range(0,no_of_col):
  name = input("Enter Colunm Names in Order {}: ".format(p))
  c.append(name)

# Calling DataFrame constructor on list
# with indices and columns specified
df = pd.DataFrame(c_main, index = None, columns = c)
df
```

**Downloading DataFrame as .csv**

**CSV File should be visible in Files section in Colab**

**If not, close the Files section and open again**

```python
file_name = input("Enter File Name (don't include extention): ") + ".csv"

# saving the dataframe
df.to_csv(file_name, index=False)
```
